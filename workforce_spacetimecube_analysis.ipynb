This section demonstrates Boolean Nasking by identifying any industries in the space time cube from above that are above the expected share of the workforce (namely, they exceed 1/13th of the workforce in their year)

```python
spacetimecube
```

Total employed in each industry (rows) each county (cols)

The totalworkforarr sums the cols of the spacetime cube. This returns the total employed in each industry, in each county, per year. The rows show the year totals, the colums show the county. 

```python
totalworkforce_arr = spacetimecube.sum(1)
totalworkforce_arr
```

Each value is divided by 13 since each value shows a workforce total in a county. Since there are 13 industries, each workforce total for the county should be evenly distributed across all industries. 

```python
thirteenth = totalworkforce_arr / 13
thirteenth
```

Each row in the preview array, thirteenth, shows 1/13 of a counties' workforce. In order to compare these numbers to the actual number of people employed in an industry in each county across all years, the rows must be repeated 13 times to check against all 13 industries. Each row is repeated 13 times and placed in a 3D array, thirteenthcube. 

```python
y2010 = np.array([thirteenth[0]] * 13)
y2011 = np.array([thirteenth[1]] * 13)
y2012 = np.array([thirteenth[2]] * 13)
y2013 = np.array([thirteenth[3]] * 13)
y2014 = np.array([thirteenth[4]] * 13)
y2015 = np.array([thirteenth[5]] * 13)
y2016 = np.array([thirteenth[6]] * 13)
y2017 = np.array([thirteenth[7]] * 13)
y2018 = np.array([thirteenth[8]] * 13)
y2019 = np.array([thirteenth[9]] * 13)
y2020 = np.array([thirteenth[10]] * 13)
y2021 = np.array([thirteenth[11]] * 13)

thirteenthcube = np.array([y2010, y2011, y2012, y2013, y2014, y2015, y2016, y2017, y2018, y2019,
            y2020, y2021])

thirteenthcube
```

A boolean array shows where a counties' workforce in a specific industry exceeds 1/13th of the projected workedforce as boolean true/false (false, does not exceed). 

```python
booleanarray = spacetimecube > thirteenthcube
booleanarray 
```

Boolean masking returns all values, which are workforce industry totals per country across 12 years, which exceed the projected workforce. 

```python
spacetimecube[spacetimecube > thirteenthcube]
```
